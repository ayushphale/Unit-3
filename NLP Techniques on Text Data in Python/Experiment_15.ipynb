```python
#Tokenizatiom
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
text = "Natural language processing is interesting"
tokens= word_tokenize(text)
print(tokens)
```

```python
#Sentence tokenization
from nltk.tokenize import sent_tokenize
text= "Python is easy. It is widely used in data science. Students like it a lot"
sentences= sent_tokenize(text)
print(sentences)
```

```python
#Stop word Removal
#Stops words such as is, the , end , and , in are removed
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
filtered_words = [w for w in words if w.lower() not in stop_words]
print(filtered_words)
```

```python
#Stemming
#Stemming reduces word to their root form.
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words=["Playing","Studies","Making","Playable"]
for w in words:
  print(stemmer.stem(w))
```

```python
#Lemmatization
#Lemmatization converts words to their dictionary form.
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("better"))
print(lemmatizer.lemmatize("studies"))
print(lemmatizer.lemmatize("making"))

```

```python
#Part-of-Speech(POS)Tagging
#POS tagging indentifies grammatical roles of words
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
words = word_tokenize(text)
pos_tags = pos_tag(words)
print(pos_tags)
```

```python
#Word Frequency Count
from nltk.probability import FreqDist
words = word_tokenize(text)
fd = FreqDist(words)
print(fd.most_common())
```
